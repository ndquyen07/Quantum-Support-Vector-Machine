```python
import numpy as np
import pickle

# Load tqfm models from results_tqfm (Digits dataset - EfficientSU2)
print("Loading TQFM models for Digits dataset - EfficientSU2 ansatz...")

with open("../results_tqfm/tqfm_digits_depth1_ansatzEfficientSU2_iter100000_20251014_232509.pkl", 'rb') as f:
    tqfm1_eff = pickle.load(f)
        
with open("../results_tqfm/tqfm_digits_depth2_ansatzEfficientSU2_iter100000_20251015_034415.pkl", 'rb') as f:
    tqfm2_eff = pickle.load(f)

with open("../results_tqfm/tqfm_digits_depth3_ansatzEfficientSU2_iter100000_20251015_090929.pkl", 'rb') as f:
    tqfm3_eff = pickle.load(f)
    
# Load kernel matrices from results_tqfm
print("\nLoading kernel matrices...")

data1 = np.load("../results_tqfm/kernels_digits_depth1_ansatzEfficientSU2_iter100000_20251014_232509.npz")
kernel_train_before1_eff = data1['kernel_before']
kernel_train_after1_eff = data1['kernel_after']
kernel_val1_eff = data1['kernel_val']
kernel_test1_eff = data1['kernel_test']


data2 = np.load("../results_tqfm/kernels_digits_depth2_ansatzEfficientSU2_iter100000_20251015_034415.npz")
kernel_train_before2_eff = data2['kernel_before']
kernel_train_after2_eff = data2['kernel_after']
kernel_val2_eff = data2['kernel_val']
kernel_test2_eff = data2['kernel_test']

data3 = np.load("../results_tqfm/kernels_digits_depth3_ansatzEfficientSU2_iter100000_20251015_090929.npz")
kernel_train_before3_eff = data3['kernel_before']
kernel_train_after3_eff = data3['kernel_after']
kernel_val3_eff = data3['kernel_val']
kernel_test3_eff = data3['kernel_test']

```

```python
print(f"tqfm1.optimal_value: {tqfm1_eff.optimal_value}")
print(f"tqfm2.optimal_value: {tqfm2_eff.optimal_value}")
print(f"tqfm3.optimal_value: {tqfm3_eff.optimal_value}")
```

```python
# draw 3 loss curves in one plot
import matplotlib.pyplot as plt
plt.plot(tqfm1_eff.loss_history, label='EfficientSU2 (depth=1)')
plt.plot(tqfm2_eff.loss_history, label='EfficientSU2 (depth=2)')
plt.plot(tqfm3_eff.loss_history, label='EfficientSU2 (depth=3)')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.legend()
```

```python
from src.utils import class_similarity, calculate_accuracy
import pandas as pd

# Define quantum kernels for all ansatzes and depths
quantum_kernels = {
	"TwoLocal (depth=1)": kernel_train_after1_two,
	"TwoLocal (depth=2)": kernel_train_after2_two,
	"TwoLocal (depth=3)": kernel_train_after3_two,
	"RealAmplitudes (depth=1)": kernel_train_after1_rea,
	"RealAmplitudes (depth=2)": kernel_train_after2_rea,
	"RealAmplitudes (depth=3)": kernel_train_after3_rea,
	"EfficientSU2 (depth=1)": kernel_train_after1_eff,
	"EfficientSU2 (depth=2)": kernel_train_after2_eff,
	"EfficientSU2 (depth=3)": kernel_train_after3_eff
}

val_kernels = {
	"TwoLocal (depth=1)": kernel_val1_two,
	"TwoLocal (depth=2)": kernel_val2_two,
	"TwoLocal (depth=3)": kernel_val3_two,
	"RealAmplitudes (depth=1)": kernel_val1_rea,
	"RealAmplitudes (depth=2)": kernel_val2_rea,
	"RealAmplitudes (depth=3)": kernel_val3_rea,
	"EfficientSU2 (depth=1)": kernel_val1_eff,
	"EfficientSU2 (depth=2)": kernel_val2_eff,
	"EfficientSU2 (depth=3)": kernel_val3_eff
}

test_kernels = {
	"TwoLocal (depth=1)": kernel_test1_two,
	"TwoLocal (depth=2)": kernel_test2_two,
	"TwoLocal (depth=3)": kernel_test3_two,
	"RealAmplitudes (depth=1)": kernel_test1_rea,
	"RealAmplitudes (depth=2)": kernel_test2_rea,
	"RealAmplitudes (depth=3)": kernel_test3_rea,
	"EfficientSU2 (depth=1)": kernel_test1_eff,
	"EfficientSU2 (depth=2)": kernel_test2_eff,
	"EfficientSU2 (depth=3)": kernel_test3_eff
}

# Calculate losses for quantum kernels
losses = {
	'TwoLocal (depth=1)': tqfm1_two.optimal_value,
	'TwoLocal (depth=2)': tqfm2_two.optimal_value,
	'TwoLocal (depth=3)': tqfm3_two.optimal_value,
	'RealAmplitudes (depth=1)': tqfm1_rea.optimal_value,
	'RealAmplitudes (depth=2)': tqfm2_rea.optimal_value,
	'RealAmplitudes (depth=3)': tqfm3_rea.optimal_value,
	'EfficientSU2 (depth=1)': tqfm1_eff.optimal_value,
	'EfficientSU2 (depth=2)': tqfm2_eff.optimal_value,
	'EfficientSU2 (depth=3)': tqfm3_eff.optimal_value
}

# Calculate separability ratios for quantum kernels
separability_ratios = {}
for name, K in quantum_kernels.items():
	w0, w1, b, ratio = class_similarity(K, y_train)
	separability_ratios[name] = ratio

# Calculate validation and test accuracies for quantum kernels
val_accuracies = {}
test_accuracies = {}

for name in quantum_kernels.keys():
	K_train = quantum_kernels[name]
	K_val = val_kernels[name]
	K_test = test_kernels[name]
	
	# Get validation accuracy, test accuracy, and best C in one call
	val_acc, test_acc, best_c = calculate_accuracy(
		K_train, K_val, K_test, y_train, y_val, y_test
	)
	
	val_accuracies[name] = val_acc
	test_accuracies[name] = test_acc

# Create comprehensive DataFrame with separability ratios, validation accuracy, test accuracy, and losses
df_results = pd.DataFrame(index=['TwoLocal (d=1)', 'TwoLocal (d=2)', 'TwoLocal (d=3)',
								'RealAmplitudes (d=1)', 'RealAmplitudes (d=2)', 'RealAmplitudes (d=3)',
								'EfficientSU2 (d=1)', 'EfficientSU2 (d=2)', 'EfficientSU2 (d=3)'],
						 columns=['Separability Ratio', 'Validation Accuracy', 'Test Accuracy', 'Loss'])

# Fill the DataFrame - Now with 4 values for 4 columns
df_results.loc['TwoLocal (d=1)'] = [separability_ratios['TwoLocal (depth=1)'], val_accuracies['TwoLocal (depth=1)'], test_accuracies['TwoLocal (depth=1)'], losses['TwoLocal (depth=1)']]
df_results.loc['TwoLocal (d=2)'] = [separability_ratios['TwoLocal (depth=2)'], val_accuracies['TwoLocal (depth=2)'], test_accuracies['TwoLocal (depth=2)'], losses['TwoLocal (depth=2)']]
df_results.loc['TwoLocal (d=3)'] = [separability_ratios['TwoLocal (depth=3)'], val_accuracies['TwoLocal (depth=3)'], test_accuracies['TwoLocal (depth=3)'], losses['TwoLocal (depth=3)']]
df_results.loc['RealAmplitudes (d=1)'] = [separability_ratios['RealAmplitudes (depth=1)'], val_accuracies['RealAmplitudes (depth=1)'], test_accuracies['RealAmplitudes (depth=1)'], losses['RealAmplitudes (depth=1)']]
df_results.loc['RealAmplitudes (d=2)'] = [separability_ratios['RealAmplitudes (depth=2)'], val_accuracies['RealAmplitudes (depth=2)'], test_accuracies['RealAmplitudes (depth=2)'], losses['RealAmplitudes (depth=2)']]
df_results.loc['RealAmplitudes (d=3)'] = [separability_ratios['RealAmplitudes (depth=3)'], val_accuracies['RealAmplitudes (depth=3)'], test_accuracies['RealAmplitudes (depth=3)'], losses['RealAmplitudes (depth=3)']]
df_results.loc['EfficientSU2 (d=1)'] = [separability_ratios['EfficientSU2 (depth=1)'], val_accuracies['EfficientSU2 (depth=1)'], test_accuracies['EfficientSU2 (depth=1)'], losses['EfficientSU2 (depth=1)']]
df_results.loc['EfficientSU2 (d=2)'] = [separability_ratios['EfficientSU2 (depth=2)'], val_accuracies['EfficientSU2 (depth=2)'], test_accuracies['EfficientSU2 (depth=2)'], losses['EfficientSU2 (depth=2)']]
df_results.loc['EfficientSU2 (d=3)'] = [separability_ratios['EfficientSU2 (depth=3)'], val_accuracies['EfficientSU2 (depth=3)'], test_accuracies['EfficientSU2 (depth=3)'], losses['EfficientSU2 (depth=3)']]


print(df_results)
print("="*50)
```

```python
from sklearn.metrics.pairwise import rbf_kernel

rbf_K_train = rbf_kernel(X_train)
rbf_K_val = rbf_kernel(X_val, X_train)
rbf_K_test = rbf_kernel(X_test, X_train)

classical_val_acc, classical_test_acc, _ = calculate_accuracy(
    rbf_K_train, rbf_K_val, rbf_K_test,
    y_train, y_val, y_test
)
print("Val acc | Test acc")
print(f"{classical_val_acc} | {classical_test_acc}")
```

```python
# Create a line chart showing validation accuracy vs depth for different anstze
depths = [1, 2, 3]

# Extract validation accuracies for each ansatz
twolocal_val_acc = [val_accuracies['TwoLocal (depth=1)'], val_accuracies['TwoLocal (depth=2)'], val_accuracies['TwoLocal (depth=3)']]
realamp_val_acc = [val_accuracies['RealAmplitudes (depth=1)'], val_accuracies['RealAmplitudes (depth=2)'], val_accuracies['RealAmplitudes (depth=3)']]
efficientsu2_val_acc = [val_accuracies['EfficientSU2 (depth=1)'], val_accuracies['EfficientSU2 (depth=2)'], val_accuracies['EfficientSU2 (depth=3)']]

# Extract test accuracies for each ansatz
twolocal_test_acc = [test_accuracies['TwoLocal (depth=1)'], test_accuracies['TwoLocal (depth=2)'], test_accuracies['TwoLocal (depth=3)']]
realamp_test_acc = [test_accuracies['RealAmplitudes (depth=1)'], test_accuracies['RealAmplitudes (depth=2)'], test_accuracies['RealAmplitudes (depth=3)']]
efficientsu2_test_acc = [test_accuracies['EfficientSU2 (depth=1)'], test_accuracies['EfficientSU2 (depth=2)'], test_accuracies['EfficientSU2 (depth=3)']]

# Create subplot for validation and test accuracies
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Validation accuracy plot
ax1.plot(depths, twolocal_val_acc, 'o-', label='TwoLocal', linewidth=2, markersize=8)
ax1.plot(depths, realamp_val_acc, 's-', label='RealAmplitudes', linewidth=2, markersize=8)
ax1.plot(depths, efficientsu2_val_acc, '^-', label='EfficientSU2', linewidth=2, markersize=8)
ax1.set_xlabel('Depth')
ax1.set_ylabel('Validation Accuracy')
ax1.set_title('Validation Accuracy vs Circuit Depth')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_xticks(depths)

# Test accuracy plot
ax2.plot(depths, twolocal_test_acc, 'o-', label='TwoLocal', linewidth=2, markersize=8)
ax2.plot(depths, realamp_test_acc, 's-', label='RealAmplitudes', linewidth=2, markersize=8)
ax2.plot(depths, efficientsu2_test_acc, '^-', label='EfficientSU2', linewidth=2, markersize=8)
ax2.set_xlabel('Depth')
ax2.set_ylabel('Test Accuracy')
ax2.set_title('Test Accuracy vs Circuit Depth')
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_xticks(depths)

plt.tight_layout()
plt.show()
```
