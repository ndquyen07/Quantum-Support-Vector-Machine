## 1. Load data

## 3. Quantum Support Vector Machine

```python
from qiskit_algorithms.optimizers import COBYLA, L_BFGS_B, SLSQP, SPSA

cobyla = COBYLA(
    maxiter=5000,
    tol=1e-20,
    rhobeg=0.5,
    options={"rhoend": 1e-20}
)
lbfgsb = L_BFGS_B(maxiter=1000)
slsqp = SLSQP(maxiter=1000)
spsa = SPSA(maxiter=5000)
```

```python
from src.qsvm_test import QSVC


qsvc = QSVC(C=10.0, gamma=1.0, depth=2, type_ansatz="EfficientSU2")
qsvc.fit(X=X_train,
	y=y_train,
	kernel_matrix=kernel,
	theta_optimal=tqfm.optimal_params,
	parametrized_circuit=tqfm.circuit,
    optimizer=cobyla
	)

accuracy = qsvc.score(X_test, y_test)
print(f"Test accuracy: {accuracy * 100:.2f}%")
```
